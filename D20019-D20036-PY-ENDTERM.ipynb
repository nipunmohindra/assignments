```python
def histogram(data):    #defining the function 
    for i in data.columns:     #iterating in the dataset
        if data[i].dtype =="int64" or data[i].dtype=="float64":        #conflicting our iteration for numeric variables
            plt.hist(data[i])  #plotting histograms 
            plt.title(f"histogram of {i}")      #this takes the title as the column name for which the histogram is made
            plt.show()
            
    return plt.show()
```

#### 2. Make an improvement on the function you have created in 1. Create a function names ‘Graphs’ which will take a dataset as input and return histograms and boxplots for all the numerical variables and bar plots for all categorical variables. (10 marks)
Graphs(data)
-> returns histograms and boxplots for all numerical variables and bar plot for all categorical variables in data.

```python
def graphs(data):   #defining the function 
    for i in data.columns:    #iterating in the dataset
        if data[i].dtype =="int64" or data[i].dtype=="float64":   #conflicting our iteration for numeric variables
            plt.hist(data[i])   #plotting histograms
            plt.title(f"histogram of {i}")  #this takes the title as the column name for which the histogram is made
            plt.show()
            plt.boxplot(data[i])    #plotting boxplots
            plt.title(f"boxplot of {i}")
            plt.show()
        else:
            plt.bar(data[i].unique(),data[i].value_counts())   #using else for categorical data and making barplots for them
            plt.title(f"barplot of {i}")

    return plt.show()
```

#### 3. Often, we are not required to plot the graphs for all the variables in our dataset. Add an additional argument to the function in 2 named ‘var’. This will take a list containing the variable index and return the graphs for only those variables. By default, (i.e. if the list of variable index is not provided), then it must return the required graphs for all the variables in data. (10 marks) 
Example: Graphs(data, var=c(1,3,4)) -> Will generate the graphics for only the variables 1,3 & 4 in the data.

```python
def graph1(data,var):   #adding var to previous function 
    for i in var        #used iloc to index label of the dataframe and confined to numeric variables
        if data.iloc[:,i].dtype == "int64" or data.iloc[:,i].dtype =="float64":  
            plt.hist(data.iloc[:,i])  #plotted histogram
            plt.title(f"histogram of {data.iloc[:,i].name}")   #this takes the title as the column name for which the histogram is made
            plt.show()   
            plt.boxplot(data.iloc[:,i])    #plotted boxplot
            plt.title(f"boxplot of {data.iloc[:,i].name}")
            plt.show()
        else:
            plt.bar(data.iloc[:,i].unique(),data.iloc[:,i].value_counts())  #categorical variables and its bar plot
            plt.title(f"barplot of {data.iloc[:,i].name}")
    return plt.show()
```

#### 4. Sometimes, we do not want to mess up our working directories with so many image files. Create an additional argument for the function "dir" (directory), such that the function exports all the files to that specified folder (which need not necessarily be your working directory). (10 marks)
Example:
Graphs(data, Variable=c(1,3,4), dir=".../Praxis/Graphs")
-> will generate all the necessary graphics for the variables 1, 3
and 4 in the specified location in your system which is ".../Praxis/
Graphs"

```python
def Graphs2(data,var,dir):
    data_new=data.iloc[:,var]    #slicing the data here
    
    numeric_data=data_new.select_dtypes(include=[np.number])   #choosing numeric variables
    categorical_data=data_new.select_dtypes(exclude=[np.number])   #choosing categorical variables
    n=numeric_data.columns   #n represents numeric data columns
    c=categorical_data.columns  #c represents categoricl data columns 
    for i in list(n):   #running a loop for numerical var and plotting distribution plot
        fig=plt.figure(figsize=(8,10))
        sns.distplot(numeric_data[i])
        plt.savefig(dir+i+'_Hist.png')   #saving the image as .png
    
    for j in list(n):
        fig=plt.figure(figsize=(10,12))   #running the loop for numeric var and plotting boxplot
        sns.boxplot(numeric_data[j])
        plt.savefig(dir+j+'_Box.png')     #saving image  as .png
    
    for k in list(c):                     ##running the loop for categorical var and plotting barplot
        fig=plt.figure(figsize(8,10))
        sns.countplot(x=categorical_data[k],data=categorical_data)
        plt.savefig(dir+k+'_Bar.png')      # #saving image  as .png
```

#### 6. New Improvements: Design a new function which you feel can help you a lot in your EDA process. (20 marks)

There are many ways to deal with missing values. If you decide to view/check the columns with too many missing values (over a certain threshold you specify), you can use this function to accomplish the task :

```python
def check(df):
    
    #takes df
    #checks nulls
    
    if df.isnull().sum().sum() > 0:        #defining the if condition 
        mask_total = df.isnull().sum().sort_values(ascending=False) #sorting values
        total = mask_total[mask_total > 0]

        mask_percent = df.isnull().mean().sort_values(ascending=False) 
        percent = mask_percent[mask_percent > 0] 

        missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])  #concatinating
    
        print(f'Total and Percentage of NaN:\n {missing_data}')  #prints total and percentage of the missing values for each column
    else: 
        print('No NaN found.')
```

```python
check(ef)
```

```python
def view (df, missing_percent):
    
    #checks which columns have over specified percentage of missing values
    #takes df, missing percentage
    #returns columns as a list
    
    mask_percent = df.isnull().mean()
    series = mask_percent[mask_percent > missing_percent]
    columns = series.index.to_list()
    print(columns) 
    return columns
```
